```python
import pandas as pd
import numpy as np
```

```python
#读入原始数据并保存
raw_data_dir = '/home/laiminzhi/wenbin/DL_stock_combo/data/xy_data/'
data = 'xy_data_new.h5'
output_dir = '/home/laiminzhi/data/feature_data_700/'

raw_data = pd.read_hdf(raw_data_dir+data)
```

```python
#把rawdata转换为一个个pivot table存到output dir下，名字为列名.pkl
for i in raw_data.columns:
    pivot = raw_data.pivot_table(index='date', columns='code', values=i)
    pivot.to_pickle(output_dir+i+'.pkl')
    print(i+' done')
```
